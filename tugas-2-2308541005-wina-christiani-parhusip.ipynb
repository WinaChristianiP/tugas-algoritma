```python
#2.7
print("--Menghitung Luas Permukaan Balok--")
#input data
panjang_05 = int(input("Panjang balok : "))
lebar_05 = int(input("Lebar balok : "))
tinggi_05 = int(input("Tinggi balok : "))
#menghitung luas
luas_permukaan_balok_05 = 2 * ((panjang_05 * lebar_05) + (lebar_05 * tinggi_05) + (panjang_05 * tinggi_05))
print("Maka Luas permukaan baloknya adalah " + str(luas_permukaan_balok_05 ) + "cm")
```
