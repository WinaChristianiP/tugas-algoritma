```python
import numpy as np
import math

# Diketahui x yang memenuhi persamaan x**2 + x - 1= 0
# Menentukan nilai n yang memenuhi x**2024-x**2025 = (x(1-x))**n

# Cari Nilai x1
x1_05 = (-1 + math.sqrt(5)) / 2
print("Nilai x adalah {}".format(x1_05))

# Hitung bagian kiri persamaan yaitu x**2024-x**2025
# x**2024 * (1-x)
kiri_05 = 2024 * math.log(x1_05) + math.log(1 - x1_05)

# Hitung bagian kanan persamaan
kanan_05 = x1_05 * (1 - x1_05)

# Hitung nilai n
n_05 = kiri_05 / math.log(kanan_05)

print(f"Diperoleh Nilai n yang memenuhi: {n_05}")
```

```python
import numpy as np
import math
import cmath

# Diketahui x yang memenuhi persamaan x**2 + x - 1= 0
# Menentukan nilai n yang memenuhi x**2024-x**2025 = (x(1-x))**n

# Cari Nilai x2
x2_05 = (-1 - math.sqrt(5)) / 2
print("Nilai x adalah {}".format(x2_05))

# Hitung bagian kiri persamaan yaitu x**2024-x**2025
# x**2024 * (1-x)
kiri_05 = 2024 * cmath.log(x2_05) + math.log(1 - x2_05)

# Hitung bagian kanan persamaan
kanan_05 = x2_05 * (1 - x2_05)

# Hitung nilai n
n_05 = kiri_05 / math.log(np.abs(kanan_05))

print(f"Diperoleh Nilai n yang memenuhi: {n_05.real}")
```
