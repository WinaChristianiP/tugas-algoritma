```python
#tipe if
#tipe rumah berdasarkan luas tertentu(dengan satuan meter)
panjang_05 = input("Panjang rumah adalah : " )
lebar_05 = input("Lebar rumah adalah : " )
luas_rumah_05 = int(panjang_05) * int(lebar_05)
print("Diketahui luas rumah adalah " + str(luas_rumah_05) + " meter kuadrat")
if luas_rumah_05 >= 300:
    print("Rumah Anda adalah tipe A1")
elif luas_rumah_05 >= 250:
    print("Rumah Anda adalah tipe A2")
elif luas_rumah_05 >= 200:
    print("Rumah Anda adalah tipe B1")
elif luas_rumah_05 >= 150:
    print("Rumah Anda adalah tipe B2")
elif luas_rumah_05 >= 70:
    print("Rumah Anda adalah tipe B3")
```

```python
#merubah celcius ke reamur ataupun sebaliknya dari reamur ke celcius
diketahui_suhu_05 = float(input("Berapakah suhu yang anda ketahui? ")) 
celcius_atau_reamur_05 = input("Satuan Suhu yang anda ketahui dalam celcius atau reamur? ")

if celcius_atau_reamur_05.lower() == 'celcius':
    suhu_05 = diketahui_suhu_05*4/5
    print("Suhu dalam satuan reamur adalah " + str(suhu_05))
elif celcius_atau_reamur_05.lower () == 'reamur':
    suhu_05 = diketahui_suhu_05*5/4
    print("Suhu dalam satuan celcius adalah " + str(suhu_05))
```
