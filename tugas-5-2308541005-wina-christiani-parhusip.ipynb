```python
#tipe 2.19
#Akan dicari Jarak, Kecepatan dan Waktu
print ("Akan dicari salah satu dari Jarak, Kecepatan dan Waktu sesuai data yang anda miliki")
pertanyaan_05 = input("Apa yang ingin anda hitung?")
if pertanyaan_05 == "Jarak":
    kecepatan_05 = input('Berapa kecepatan yang dipakai?')
    waktu_05 = input('Berapakah waktu yang ditempuh? ')
    jarak_05 = int(kecepatan_05) * int(waktu_05)
    print(f"Jaraknya adalah {jarak_05} km")
elif pertanyaan_05 == "Kecepatan":
    jarak_05 = input('Berapakah jarak yang ditempuh? ')
    waktu_05 = input('Berapa lama waktu yang ditempuh? ')
    kecepatan_05 = int(jarak_05) / int(waktu_05)
    print(f"Maka kecepatannya adalah {kecepatan_05} km/jam")
elif pertanyaan_05 == "Waktu":
    jarak_05 = input('Berapakah jarak yang ditempuh? ')
    kecepatan_05 = input("Berapa kecepatan yang dipakai?")
    waktu_05 = int(jarak_05) / int(kecepatan_05)
    print(f"Waktu tempuh adalah selama adalah {waktu_05} jam")
```
