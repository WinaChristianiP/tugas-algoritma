```python
import numpy as np
import matplotlib.pyplot as plt

print('----------PERSAMAAN GARIS----------')
#Akan dicari Persamaan garis pertama yang melewati titik (a,b) dan (c,d)
print("-------Persamaan garis pertama-------")
a_05 = int(input("Masukkan nilai a: "))   #input nilai titik(a,b) dan (c,d)
b_05 = int(input("Masukkan nilai b: "))
c_05 = int(input("Masukkan nilai c: "))
d_05 = int(input("Masukkan nilai d: "))
m1_05 = (d_05 - b_05) / (c_05 - a_05)    #Mencari gradien yang pertama
print(f'Gradien garisnya adalah {m1_05}')

print(" ")

#Akan dicari Persamaan garis kedua yang melewati titik (e,f) dan (g,h)
print("-------Persamaan garis kedua-------")
e_05 = int(input("Masukkan nilai e: "))   #input nilai titik(e,f) dan (g,h)
f_05 = int(input("Masukkan nilai f: "))
g_05 = int(input("Masukkan nilai g: "))
h_05 = int(input("Masukkan nilai h: "))
m2_05 = (h_05 - f_05) / (g_05 - e_05)    #mencari gradien yang kedua
print(f'Gradien garisnya adalah {m2_05}')

#Maka akan diperoleh grafik
x_05= np.arange(-10,10,0.01)
plt.plot(x_05,m1_05 * (x_05 - a_05) + b_05, color = "blue", linewidth = 2)    #Plot garis pertama
plt.plot(x_05,m2_05 * (x_05 - e_05) + f_05, color = 'yellow', linewidth = 2)  #plot garis kedua

#hasil
plt.title("GRAFIK")
print("Diperoleh gambar grafik seperti dibawah ini")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("<<<===Persamaan Garis===>>>")
#Persamaan garis pertama melewati titk a,b dan c,d
a_05 = -1 
b_05 = 4
c_05 = 3
d_05 = -8
m1_05= (d_05 - b_05) / (c_05 - a_05) #Mencari gradien garis ke 1
print("Gradien garis pertama adalah{}".format(m1_05))
print("Persamaan garis pertamaa adalah y={}x+{} ".format(m1_05,-m1_05 * a_05 + b_05))
#y1_05= m1_05 * (x - a_05) + b_05

#Persamaan garis kedua melewati titik e,f dan g,h
e_05 = 12
f_05 = -25
g_05 = 9
h_05 = -4
m2_05 = (h_05 - f_05) / (g_05 - e_05) #Mencari gradien garis ke 2
print("Gradien garis kedua  adalah{}".format(m2_05))
print("Persamaan garis kedua adalah y={}x+{} ".format(m2_05,-m2_05 * e_05 + f_05))
#y2_05 = m2_05 * (x - e_05) + f_05

if m1_05==m2_05:
    print("Gradiennya sama, jadi kemungkinannya ga ada solusi atau solusinya banyak")
else:
    #Titik potong kedua garis pada (tp_x,tp_y)
    tp_x_05 = (m1_05 * a_05 - b_05- m2_05 * e_05 + f_05) / (m1_05 - m2_05)
    tp_y_05 = m2_05 * (tp_x_05 - e_05) + f_05
    
    print("Titik potong untuk adalah({},{})".format(tp_x_05,tp_y_05))
    
    x_05= np.arange(tp_x_05 - 10,tp_x_05 + 10,0.01)
    plt.figure(figsize = (4,4),dpi = 300)
    
    plt.plot(x_05,m1_05 * (x_05 - a_05) + b_05, color ='Red',label = 'y={}x+{}'.format(m1_05,-m1_05 * a_05 + b_05))
    plt.plot(x_05,m2_05 * (x_05 - e_05) + f_05, color ='Yellow',label = 'y={}x+{}'.format(m2_05,-m2_05 * e_05 + f_05))
    plt.plot(tp_x_05,tp_y_05, color= "Blue",marker='1',markersize = 4)
    plt.legend(loc='best')
    plt.show()
```
