```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

#Mulai
#Menggunakan nilai a,b, dan c dan akar-akar
# ax**2 + bx + c = 0
print("====Program Persamaan Kuadrat====")
print("Hai selamat datang di pemogramaan kali ini")
print("Maskkan nilai a, b, dan c dari persamaan yang diketahui")
a_05= float(input("Masukkan nilai a : "))
b_05= float(input("Masukkan nilai b : "))
c_05= float(input("masukkan nilai c : "))
if a_05 != 0 :
    print(f"Kita peroleh Nilai a = {a_05}")
    print(" ")
    print("Mari lanjutkan mencari bentuk kurva")
    print("Dari nilai a, Apakah kamu tau kira kira kurva atau grafik anda terbuka ke arah manna?")
    #Selanjutnya kita akan melihat apakah kurvanya terbuka keatas atau kebawah
    if a_05 > 0 :
        print("Yaps, Kurva anda terbuka keatas")
        print("Selanjutnya ayo cari diskrminan nyaa!!!")
    elif a_05 < 0 :
        print("Yaps, Kurva anda terbuka kebawah")
        print("Selanjutnya ayo cari diskrminan nyaa!!!")
    #selanjutnya cari nilai diskriminan
    disk_05 = (b_05*b_05)- 4*a_05*c_05
    print(f"Nilai diskriminan dari fungsi ini kita peroleh {disk_05}")
    print("Eitsss, belum selesai, ayo cari jenis jenis akar persamaan kuadrat berdasrkan nilai diskriman yang telah diperoleh")
    if disk_05 > 0 :
        print("Karena nilai diskriminan nya > 0, maka Akar akarnya adalah akar real berbeda ")
        print("Selanjutnya ayo cari akar-akarnya") #Selanjutnya mari kita cari akar-akarnya
        x1_05 = (-b_05 + cmath.sqrt(disk_05)) / (2*a_05)
        x2_05 = (-b_05 - cmath.sqrt(disk_05)) / (2*a_05)
        print(f"maka akar x1 adalah {x1_05} dan akar x2 adalah {x2_05}")
    elif disk_05 == 0 :
        print("Karena nilai diskrimanannya = 0, Nanti akar-akarnya akan kembar")
        print("Ayo cari akar kembar tersebut!") #Selanjutnya mari kita cari akar-akarnya
        x1_05 = (-b_05 + cmath.sqrt(disk_05)) / (2*a_05)
        x2_05 = x1_05
        print(f"Akar kembar nya adalah {x1_05}")
    elif disk_05 < 0 :
        print("karena nilai diskriminan nya < 0,berarti nilai akarnya nanti imajiner")
        print("Ayo tetap cari nilai akarnya!!!")
        x1_05 = (-b_05 + cmath.sqrt(disk_05)) / (2*a_05)
        x2_05 = (-b_05 - cmath.sqrt(disk_05)) / (2*a_05)
        print(f"maka akar x1 adalah {x1_05} dan akar x2 adalah {x2_05}")
        
    #Selanjutnya akan dicari titik puncak dan sumbu simetri
    tpcx_05 = -b_05 / 2*a_05
    tpcy_05 = -disk_05 / 4*a_05
    print(f"Maka titik puncaknya adalah {tpcx_05},{tpcy_05}.")
    sbsm_05 = -b_05 / 2*a_05
    print(f"Sumbu simetri nya adalah {sbsm_05}")
    #mencari posisi titik puncak
    if tpcx_05 > 0:
        print("posisi titik puncak di kanan sumbu y")
    elif tpcx_05 == 0 :
        print ("Posisi Titik puncak berada pada sumbu y")
    elif tpcx_05 < 0:
        print ("Posisi Titik puncak berada di kiri sumbu y")
    #melihat kondisi parabola memotong sumbu y
    if c_05 > 0 :
        print("parabola memotong sumbu y positif")
    elif c_05 == 0:
        print("parabola memotong di pusat (0,0)")
    elif c_05 < 0 :
        print("parabola memotong sumbu y negatif")
    #membuat grafik
    #Menentukan rentang nilai x
    x_05 = np.linspace(tpcx_05 - 100, tpcy_05 + 100, 500)
    y_05 = a_05*x_05**2 + b_05*x_05 + c_05
    plt.figure(figsize=(8, 8))
    plt.plot(x_05, y_05, label=f'{a_05}x^2 + {b_05}x + {c_05} = 0', color='brown')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=2)
    plt.scatter(tpcx_05, tpcy_05, color ='blue', label= 'Titik puncak')
    plt.axvline(tpcx_05, color='green', linestyle = '--', label = 'Sumbu simetris')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Kuadrat')
    plt.show()
        
elif a_05 == 0 :
    print("Oops, a tidak boleh 0")    
```
